```python
# contabilizar cuantas de la columna 4 (Categoría), se tiene de la columna 3 con iloc. Y que me muestre el valor de la columna 3 y el valor de la 4 y cuantos de la columna 4 hay de cada uno de la columna 3
# Acceder a las columnas usando iloc
columna_3 = df.iloc[:, 2]  # Columna 3 (índice 2)
columna_4 = df.iloc[:, 3]  # Columna 4 (índice 3)

# Crear un DataFrame temporal con las dos columnas
temp_df = pd.DataFrame({
    'columna_3': columna_3,
    'categoria': columna_4
})

# Contabilizar las combinaciones
conteo = temp_df.groupby(['columna_3', 'categoria']).size().reset_index(name='Cantidad')
```

```python
print("Conteo de Columna 4 por cada valor de Columna 3:")
conteo
```

```python
# También puedes ver un resumen por columna 3
resumen_por_col3 = temp_df.groupby('columna_3')['categoria'].value_counts().reset_index(name='Cantidad')
print("\nResumen detallado:")
resumen_por_col3
```

```python
# Para ver el total por cada valor de columna 3
total_por_col3 = temp_df.groupby('columna_3').size().reset_index(name='Total')
print("\nTotal por Columna 3:")
total_por_col3
```
